```python
objects = df.columns[df.dtypes == 'object']
objects
```

```python
for obj in objects:
    df[obj] = pd.Categorical(df[obj]).codes
df
```

```python
df.dtypes
```

```python
from sklearn.model_selection import train_test_split
X = df.loc[:, :'nr.employed']
y = df['y']
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2)
```

```python
y_test.sum()
```

```python
training_data_inputs = []
training_data_results = []
for i in X_train.index:
    training_data_inputs.append(np.reshape(X_train.loc[i].values, (20,1)))
    training_data_results.append(y_train.loc[i])
print(len(training_data_inputs))
train_data = zip(training_data_inputs, training_data_results)
```

```python
test_data_inputs = []
test_data_results = []
for i in X_test.index:
    test_data_inputs.append(np.reshape(X_test.loc[i].values, (20,1)))
    test_data_results.append(y_test.loc[i])
print(len(test_data_inputs))
test_data = zip(test_data_inputs, test_data_results)
```

```python
import pickle
pickle.dump(test_data, open("Saved_Objects/bank_marketing/test.p", "wb"))
pickle.dump(train_data, open("Saved_Objects/bank_marketing/train.p", "wb"))
```
