```python
import pandas as pd
df = pd.read_csv('Saved_Objects/heloc/heloc_dataset_v1.csv')
df.head()
```

```python
df.RiskPerformance = pd.Categorical(df.RiskPerformance).codes
```

```python
df.dtypes
```

```python
df.head()
```

```python
df.describe()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.loc[:,'ExternalRiskEstimate':], df.loc[:,'RiskPerformance'], test_size=0.25)
```

```python
y_test.sum()
```

```python
len(X_train)
```

```python
training_data_inputs = []
training_data_results = []
for i in X_train.index:
    training_data_inputs.append(np.reshape(X_train.loc[i].values , (23,1)))
    training_data_results.append(y_train.loc[i])
train_data = zip(training_data_inputs, training_data_results)
```

```python
test_data_inputs = []
test_data_results = []
for i in X_test.index:
    test_data_inputs.append(np.reshape(X_test.loc[i].values , (23,1)))
    test_data_results.append(y_test.loc[i])
test_data = zip(test_data_inputs, test_data_results)
```

```python
import pickle

pickle.dump(train_data, open("Saved_Objects/heloc/train.p", "wb"))
pickle.dump(test_data, open("Saved_Objects/heloc/test.p", "wb"))
```

## Train

```python
import pickle
train = pickle.load(open("Saved_Objects/heloc/train.p", "rb"))
test = pickle.load(open("Saved_Objects/heloc/test.p", "rb"))
```

```python
import network
net = network.Network([23, 15, 10, 1], return_vector=True)
net.SGD(train, 30, 10, 0.5, test_data=test)
```
