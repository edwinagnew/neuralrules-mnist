```python
import pickle
parser = pickle.load(open("data_parser.p", "rb"))
```

```python
cols = [val[0] for val in parser.values()]
cols[:10]
```

```python
data = []
for line in lines:
    s = []
    for i in range(1,len(parser.keys()) + 1):
        form = parser[i]
        s.append(int(line[int(form[1])-1:int(form[1])-1+int(form[2])]))
    data.append(s)
print(len(data))
print(len(data[1]))
```

```python
import pandas as pd
df_actual = pd.DataFrame(data, columns=cols)
df_actual
```

```python
df3 = df_actual.drop(['ICPSR study number-9574','ICPSR edition number-','ICPSR part number-1'], axis=1)
df2 = df3[df3['Time from probation sentence to arrest'] < 10]
df2
```

```python
df2.loc[28, :'Presentence investigation before sentencing']
```

```python
import numpy as np
msk = np.random.rand(len(df2)) < 0.85
train = df2[msk]
test = df2[~msk]
```

```python
train.loc[train.index[0], 'Time from probation sentence to arrest']
```

```python
print("test", test[test['Time from probation sentence to arrest'] == 0].count().values[0])
print("train", train[train['Time from probation sentence to arrest'] == 0].count().values[0])
```

```python
#For predicting probation recommendation
training_data_inputs = []
training_data_results = []
for i in train.index:
    training_data_inputs.append(np.reshape(train.loc[i, :'Status of request-receipt of rap sheet information'].values , (128,1)))
    training_data_results.append(train.loc[i, 'Probation recommended'])
train_data = zip(training_data_inputs, training_data_results)
```

```python
print(len(training_data_inputs))
print(len(training_data_results))
```

```python
#For predicting probation recommendation
test_data_inputs = []
test_data_results = []
for i in test.index:
    test_data_inputs.append(np.reshape(test.loc[i, :'Presentence investigation before sentencing'].values , (128,1)))
    test_data_results.append(test.loc[i, 'Probation recommended'])
test_data = zip(test_data_inputs, test_data_results)
```

```python
print(len(test_data_inputs))
print(len(test_data_results))
```

```python
pickle.dump(train_data, open("neuralrules-mnist/Saved_Objects/recidivism/recidivism_train.p", "wb"))
pickle.dump(test_data, open("neuralrules-mnist/Saved_Objects/recidivism/recidivism_test.p", "wb"))
```
